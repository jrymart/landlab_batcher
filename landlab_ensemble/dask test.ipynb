```python
from dask.distributed import Client, progress
client = Client(threads_per_worker=1, n_workers=3)
client
```

```python
import time
import random

def inc(x):
    time.sleep(15)
    return x + 1, x+2
```

```python
res = []
for i in range(25):
    f = client.submit(inc, i)
    res.append(f)
```

```python
j = 25
while j < 30:
    time.sleep(3)
    try:
        index = [r.status for r in res].index('finished')
        finished = res.pop(index)
        print("got %d when finished %s" % (finished.result(), finished))
        f = client.submit(inc, j)
        print("submitted %d" % j)
        j += 1
    except ValueError:
        print("no finished jobs")
```
